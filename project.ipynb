```python
import cv2
from skimage.io import imread
from matplotlib.pyplot import imshow, subplot
from rembg import remove
import numpy as np

# Load the images
input_img1 = imread("D:\\SY\\4th SEM\\Computer Vision\\Project\\IMG-20240424-WA0003.jpg")
bg_img1 = imread("D:\SY\\4th SEM\Computer Vision\Project\\bg3.jpg")


# Remove background from the input image
rembg_img1 = remove(input_img1)

# Add an alpha channel to the background image
bg_img1_with_alpha = np.dstack((bg_img1, np.ones((bg_img1.shape[0], bg_img1.shape[1]), dtype=np.uint8) * 255))

# Blend the images
alpha = 0.55  
output_img1 = cv2.addWeighted(rembg_img1, alpha, bg_img1_with_alpha, 1-alpha, 0)

imshow(bg_img1)
imshow(rembg_img1)
```

```python
import cv2
from skimage.io import imread
from matplotlib.pyplot import imshow
from rembg import remove
import numpy as np

input_img4 = imread("D:\\SY\\4th SEM\\Computer Vision\\Project\\img2.jpg")
bg_img4 = imread("D:\SY\\4th SEM\Computer Vision\Project\\bg5.jpg")

bg_img4=cv2.resize(bg_img4, (input_img4.shape[1], input_img4.shape[0]))

rembg_img4 = remove(input_img4) # Remove background

bg_img4_with_alpha = np.dstack((bg_img4, np.ones((bg_img4.shape[0], bg_img4.shape[1]), dtype=np.uint8) * 255))

# Blend the images
alpha = 1
output_img4 = cv2.addWeighted(rembg_img4, alpha, bg_img4_with_alpha, 2-alpha, 6) #6 contributes to brightness  and 2 if for opacity of the background image 
imshow(bg_img4)
imshow(rembg_img4)
```
